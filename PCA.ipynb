```python
import pandas as pd
df=pd.read_csv('audit_data.csv')
df.head()
```

```python
df.info()
```

```python
df.drop('LOCATION_ID',axis=1,inplace=True)
```

```python
df=df.dropna()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
risk=df['Risk'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.pie(risk,risk.index,autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```

```python
y=df['Risk']
x=df.drop('Risk',axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_df=scaler.fit_transform(x)

scaled_df[0:7]
```

```python
scaled_df
```

```python

scaled_df=pd.DataFrame(scaled_df,columns=scaled_df.columns)
```

```python
scaled_df.head()
```

```python
scaled_df.describe()
```

```python
from sklearn.decomposition import PCA
pca=PCA()
```

```python
pca.fit(scaled_df)
```

```python
pd.DataFrame(pca.explained_variance_ratio_)
```

```python
import numpy as np
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('no of components')
plt.ylabel('variance')
plt.title('cumsum plot')
plt.grid()
plt.show()
```

```python
pca=PCA(n_components=7)
pca.fit(scaled_df)
scaled_df_x=pca.transform(scaled_df)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
from sklearn.svm import SVC
sc=SVC()
sc.fit(x_train,y_train)
```

```python
sc.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
pred=sc.predict(x_test)
accuracy_score(y_test,pred)
```
